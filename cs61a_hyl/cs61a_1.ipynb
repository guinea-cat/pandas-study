ä½ åªèƒ½è°ƒç”¨ä»¥ä¸‹ä¸‰ä¸ªå‡½æ•°ï¼š
* $f(x) = x - 1$
* $g(x) = 2x$
* $h(x, y) = \text{concat}(x, y)$ (ä¾‹å¦‚ $h(789, 12) \to 78912$)

å®šä¹‰ï¼š
ä¸€ä¸ª â€œå°è¡¨è¾¾å¼â€ï¼ˆSmall Expressionï¼‰ æ˜¯æŒ‡ä»…åŒ…å«å‡½æ•° f, g, hï¼Œæ•°å­— 5 ä»¥åŠæ‹¬å·çš„è°ƒç”¨è¡¨è¾¾å¼ã€‚è¿™äº›å…ƒç´ å¯ä»¥é‡å¤ä½¿ç”¨ã€‚
* ç¤ºä¾‹ï¼š$h(g(5), f(f(5)))$ çš„æ±‚å€¼ç»“æœä¸º 103ã€‚

ç›®æ ‡ï¼š
æ‰¾åˆ°æ±‚å€¼ç»“æœä¸º 2024 çš„æœ€çŸ­å°è¡¨è¾¾å¼ï¼ˆå­—ç¬¦é•¿åº¦æœ€çŸ­ï¼‰ã€‚

```python
import collections # BFS æœç´¢ å±‚å±‚æ‰©æ•£ æ‰¾æœ€çŸ­è·¯å¾„ ä»5å¼€å§‹

def f(x): return x - 1
def g(x): return 2 * x
def h(x, y): 
    try:
        s = str(x) + str(y)
        if len(s) > 10: return None # é˜²æ­¢æ•°å­—è¿‡å¤§æº¢å‡º
        return int(s)
    except:
        return None

def solve():
    target = 2024
    # åˆå§‹çŠ¶æ€
    initial_val = 5
    initial_expr = "5" # è¡¨è¾¾å¼
    
    # queue å­˜æ”¾ (å½“å‰å€¼, å½“å‰è¡¨è¾¾å¼) å…ƒç»„ (5, "5")
    queue = collections.deque([(initial_val, initial_expr)])
    
    # è®°å½•æ¯ä¸ªæ•°å€¼å¯¹åº”çš„æœ€çŸ­è¡¨è¾¾å¼ å­—å…¸ è®°å½•å»è¿‡çš„åœ°æ–¹çš„æœ€çŸ­è·¯å¾„ï¼Œå³è¡¨è¾¾å¼
    # é˜²æ­¢æˆ‘ä»¬åšæ— ç”¨åŠŸï¼ˆæ¯”å¦‚ 5 -> 4 -> 5 è¿™ç§å›å¤´è·¯
    visited = {initial_val: initial_expr}
    
    # å­˜å‚¨æ‰€æœ‰å·²ç»å‘ç°çš„æ•°å­—ï¼Œç”¨äº h(x, y) çš„ç»„åˆ
    found_nums = {initial_val: initial_expr}

    while queue:
        val, expr = queue.popleft() # åŒç«¯é˜Ÿåˆ— å·¦ä¾§å¼¹å‡º

        if val == target:
            return expr

        # å€™é€‰æ“ä½œåˆ—è¡¨
        moves = []
        
        # 1. f(x)
        if val > 1:
            moves.append((f(val), f"f({expr})")) # å¦‚(4, "f(5)")
        
        # 2. g(x)
        if val < target * 2:
            moves.append((g(val), f"g({expr})"))
            
        # 3. h(x, y) æ‹¼æ¥é€»è¾‘
        # å°è¯•å°†å½“å‰ç”Ÿæˆçš„ val ä¸åº“ä¸­å·²æœ‰çš„æ‰€æœ‰æ•°å­—è¿›è¡ŒåŒå‘æ‹¼æ¥
        for other_val, other_expr in found_nums.items(): # dict.items() è¿”å›é”®å€¼å¯¹
            # h(val, other) åˆæ³•çš„ç»“æœåŠ å…¥ moves
            res_a = h(val, other_val)
            if res_a is not None and res_a <= 10000:
                moves.append((res_a, f"h({expr},{other_expr})"))
            
            # h(other, val)
            res_b = h(other_val, val)
            if res_b is not None and res_b <= 10000:
                moves.append((res_b, f"h({other_expr},{expr})"))

        for n_val, n_expr in moves:
            # å¦‚æœæ˜¯æ–°æ•°å­—ï¼Œæˆ–è€…æ‰¾åˆ°äº†æ›´çŸ­çš„å­—ç¬¦ä¸²è¡¨è¾¾å¼
            if n_val not in visited or len(n_expr) < len(visited[n_val]):
                visited[n_val] = n_expr
                found_nums[n_val] = n_expr
                queue.append((n_val, n_expr)) # å…¥é˜Ÿ

if __name__ == "__main__":
    answer = solve()
    print(f"Target: 2024")
    print(f"Shortest Expression: {answer}")
    print(f"Length: {len(answer)}")
```

```python
def is_prime(n):
    """
    å¦‚æœ n æ˜¯ç´ æ•°è¿”å› Trueï¼Œå¦åˆ™è¿”å› Falseã€‚

    >>> is_prime(10)
    False
    >>> is_prime(7)
    True
    >>> is_prime(1) # 1 ä¸æ˜¯ç´ æ•°
    False
    """
    # 1. å¤„ç†ç‰¹æ®Šæƒ…å†µï¼š1 ä¸æ˜¯ç´ æ•°
    if n <= 1:
        return False
    
    # 2. å®šä¹‰é¢å¤–çš„å˜é‡ï¼šä» 2 å¼€å§‹å°è¯•é™¤ n
    k = 2
    
    # 3. å¾ªç¯è¿‡ç¨‹ï¼šä¸€ç›´è¯•åˆ° n-1 ä¸ºæ­¢
    while k < n:
        # å¦‚æœ n èƒ½è¢« k æ•´é™¤ï¼ˆä½™æ•°ä¸º0ï¼‰ï¼Œè¯´æ˜å®ƒä¸æ˜¯ç´ æ•°
        if n % k == 0:
            return False
        # å°è¯•ä¸‹ä¸€ä¸ªæ•°å­—
        k += 1
        
    # 4. å¦‚æœå¾ªç¯è·‘å®Œäº†éƒ½æ²¡æœ‰è¿”å› Falseï¼Œè¯´æ˜æ²¡æœ‰æ‰¾åˆ°é™¤ 1 å’Œ n ä»¥å¤–çš„å› å­
    return True

# --- ä»¥ä¸‹ä¸ºæµ‹è¯•ä»£ç  ---
if __name__ == "__main__":
    # æµ‹è¯•å‡ ä¸ªä¾‹å­
    print("is_prime(10):", is_prime(10)) # åº”è¯¥è¾“å‡º False
    print("is_prime(13):", is_prime(13)) # åº”è¯¥è¾“å‡º True
    print("is_prime(1):",  is_prime(1))  # åº”è¯¥è¾“å‡º False
    print("is_prime(9):",  is_prime(9))  # åº”è¯¥è¾“å‡º False
```

æç¤ºï¼šä½ å¯ä»¥ä½¿ç”¨ // å’Œ % è¿ç®—ç¬¦å°†æ­£æ•´æ•°æ‹†åˆ†ä¸ºä¸ªä½å’Œå…¶ä»–ä½ã€‚

å»ºè®®ï¼šå…ˆå®šä¹‰ä¸€ä¸ªè¾…åŠ©å‡½æ•° has_digit(n, k)ï¼Œç”¨äºåˆ¤æ–­æ•°å­— n ä¸­æ˜¯å¦åŒ…å«æ•°å­— k

```python
def unique_digits(n):
    """Return the number of unique digits in positive integer n.

    >>> unique_digits(8675309) # All are unique
    7
    >>> unique_digits(13173131) # 1, 3, and 7
    3
    >>> unique_digits(101) # 0 and 1
    2
    """
    count = 0
    k = 0
    # éå† 0 åˆ° 9 è¿™åä¸ªå¯èƒ½çš„æ•°å­—
    while k < 10:
        if has_digit(n, k):
            count += 1
        k += 1
    return count

def has_digit(n, k):
    """Returns whether k is a digit in n.

    >>> has_digit(10, 1)
    True
    >>> has_digit(12, 7)
    False
    """
    # éå† n çš„æ¯ä¸€ä½
    while n > 0:
        last = n % 10  # å–å‡ºä¸ªä½
        if last == k:
            return True
        n = n // 10    # å»æ‰ä¸ªä½
    return False
```

```python
# Cell 1: å®šä¹‰æœ€åŸºç¡€çš„ç­–ç•¥å‡½æ•°
# è¿™æ˜¯ä¸€ä¸ªâ€œæ²‰é»˜çš„æ£‹æ‰‹â€ï¼Œå®ƒåªç®¡æ€è€ƒï¼Œä¸è¯´è¯

def two_strat(n):
    """æ°¸è¿œè¿”å› 2 çš„ç­–ç•¥"""
    return 2

print("âœ… åŸºç¡€ç­–ç•¥ two_strat å·²å®šä¹‰å®Œæˆ")
```

```python
# Cell 2: å®šä¹‰â€œåµé—¹â€çš„ç­–ç•¥åŒ…è£…å™¨
# è¿™æ˜¯ä¸€ä¸ªâ€œè§£è¯´å‘˜â€ï¼Œå®ƒæ¥æ”¶ä¸€ä¸ªç­–ç•¥ï¼ŒæŠŠå®ƒåŒ…è£…æˆä¸€ä¸ªä¼šæ‰“å°è¿‡ç¨‹çš„æ–°ç­–ç•¥

def noisy_strat(who, s):
    """
    who: ç©å®¶ç¼–å· (0 æˆ– 1)
    s:   åŸæœ¬çš„ç­–ç•¥å‡½æ•° (æ¯”å¦‚ two_strat)
    """
    def strat(n):
        # 1. è°ƒç”¨åŸæœ¬çš„ç­–ç•¥ s è·å–ä¸‹ä¸€æ­¥èµ°æ³•
        choice = s(n)
        
        # 2. ã€å¯è§†åŒ–ã€‘æ‰“å°å‡ºè¿™ä¸€æ­¥å‘ç”Ÿäº†ä»€ä¹ˆ
        # ä¸ºäº†è®©ä½ çœ‹æ¸…æ¥šï¼Œæˆ‘åŠ äº†ç®­å¤´ ->
        print(f'ğŸ“¢ ç©å®¶ {who} (å½“å‰æ€»æ•° {n}) -> é€‰æ‹©åŠ  {choice} -> æ–°æ€»æ•° {choice + n}')
        
        # 3. è¿”å›åŸæœ¬çš„å†³ç­–ç»“æœ
        return choice
        
    return strat # è¿”å›è¿™ä¸ªå…·å¤‡æ‰“å°åŠŸèƒ½çš„æ–°å‡½æ•°

print("âœ… è£…é¥°å™¨ noisy_strat å·²å®šä¹‰å®Œæˆ")
```

```python
# Cell 3: å®šä¹‰æ¸¸æˆä¸»æµç¨‹
# è¿™æ˜¯ä¸€ä¸ªâ€œè£åˆ¤â€ï¼Œè´Ÿè´£è½®æµè®©ç©å®¶è¡ŒåŠ¨ï¼Œå¹¶åˆ¤æ–­è¾“èµ¢

def play(strategy0, strategy1, goal=21):
    """
    strategy0: ç©å®¶0çš„ç­–ç•¥å‡½æ•°
    strategy1: ç©å®¶1çš„ç­–ç•¥å‡½æ•°
    goal: ç›®æ ‡åˆ†æ•° (é»˜è®¤ä¸º21)
    """
    print(f"\n{'='*20} æ¸¸æˆå¼€å§‹ (ç›®æ ‡: {goal}) {'='*20}")
    
    n = 0       # ç°åœ¨çš„æ€»æ•°
    who = 0     # å½“å‰è½®åˆ°è° (0 æˆ– 1)

    while n < goal:
        if who == 0:
            # è½®åˆ°ç©å®¶0ï¼Œæ‰§è¡Œ strategy0
            n = n + strategy0(n)
            who = 1 # æ¢äººï¼Œä¸‹å›åˆç»™ç©å®¶1
        elif who == 1:
            # è½®åˆ°ç©å®¶1ï¼Œæ‰§è¡Œ strategy1
            n = n + strategy1(n)
            who = 0 # æ¢äººï¼Œä¸‹å›åˆç»™ç©å®¶0
            
    print(f"{'='*20} æ¸¸æˆç»“æŸ {'='*20}")
    
    # æŒ‰ç…§åŸä»£ç é€»è¾‘ï¼Œæœ€å who æŒ‡å‘çš„æ˜¯â€œè¿˜æ²¡æ¥å¾—åŠåŠ¨çš„é‚£ä¸ªäººâ€
    # å¦‚æœç©å®¶0èµ°äº†ä¸€æ­¥å¯¼è‡´ n >= 21ï¼Œwho ä¼šå˜æˆ 1ï¼Œå¾ªç¯ç»“æŸã€‚
    # æ‰€ä»¥è¿”å› 1 æ„å‘³ç€ ç©å®¶0 èµ°å‡ºäº†æœ€åä¸€æ­¥ã€‚
    return who

print("âœ… æ¸¸æˆå¼•æ“ play å·²å®šä¹‰å®Œæˆ")
```

```python
# Cell 4: è¿è¡Œæ¸¸æˆ
# æˆ‘ä»¬ç°åœºç»„è£…ä¸¤ä¸ªâ€œåµé—¹çš„â€ç­–ç•¥ï¼Œå¹¶å¼€å§‹å¯¹æˆ˜

# ç»„è£…è¿‡ç¨‹ï¼š
# 1. noisy_strat(0, two_strat) -> åˆ¶é€ ä¸€ä¸ªâ€œæˆ‘æ˜¯ç©å®¶0ï¼Œæˆ‘æ¯æ¬¡åŠ 2ï¼Œå¹¶ä¸”æˆ‘ä¼šå–Šå‡ºæ¥â€çš„å‡½æ•°
# 2. noisy_strat(1, two_strat) -> åˆ¶é€ ä¸€ä¸ªâ€œæˆ‘æ˜¯ç©å®¶1ï¼Œæˆ‘æ¯æ¬¡åŠ 2ï¼Œå¹¶ä¸”æˆ‘ä¼šå–Šå‡ºæ¥â€çš„å‡½æ•°

winner = play(noisy_strat(0, two_strat), noisy_strat(1, two_strat))

print(f"\nğŸ† æœ€ç»ˆè¿”å›å€¼ (winner): {winner}")
if winner == 1:
    print("(æ„æ€æ˜¯ç©å®¶ 0 èµ°å‡ºäº†æœ€åä¸€æ­¥ï¼ŒæŠŠæ•°å­—æ¨åˆ°äº† 21 ä»¥ä¸Š)")
else:
    print("(æ„æ€æ˜¯ç©å®¶ 1 èµ°å‡ºäº†æœ€åä¸€æ­¥ï¼ŒæŠŠæ•°å­—æ¨åˆ°äº† 21 ä»¥ä¸Š)")
```

```python
# Cell 4: ä½œç”¨åŸŸæŠ¥é”™æ¼”ç¤º & Lambda åŸºç¡€

# å‡½æ•°åªèƒ½çœ‹åˆ°è‡ªå·±å†…éƒ¨çš„å˜é‡å’Œå…¨å±€å˜é‡ï¼Œçœ‹ä¸åˆ°â€œè°ƒç”¨å®ƒçš„å‡½æ•°â€é‡Œçš„å˜é‡ã€‚

# --- Part 1: ä½œç”¨åŸŸæŠ¥é”™ (Lexical Scope) ---
def g(a):
    # è¿™é‡Œè¯•å›¾è®¿é—® y
    # ä½†æ˜¯ g å®šä¹‰åœ¨å…¨å±€ï¼Œå®ƒåªèƒ½æ‰¾ 1.è‡ªå·±å†…éƒ¨çš„å˜é‡ 2.å…¨å±€å˜é‡
    # å®ƒçœ‹ä¸è§ f å†…éƒ¨çš„ y
    return a + y 

def f(x, y):
    # è™½ç„¶è¿™é‡Œå®šä¹‰äº† yï¼Œä½† g çœ‹ä¸è§
    return g(x)

print("--- å°è¯•è¿è¡Œ f(1, 2) ---")
try:
    f(1, 2)
except NameError as e:
    print(f"âŒ æŠ¥é”™äº†! {e}")
    print("è§£é‡Š: g å‡½æ•°åœ¨è¿è¡Œæ—¶æ‰¾ä¸åˆ°å˜é‡ 'y'ï¼Œå› ä¸ºå®ƒä¸åœ¨å…¨å±€ä½œç”¨åŸŸï¼Œä¹Ÿä¸åœ¨ g çš„å±€éƒ¨ä½œç”¨åŸŸã€‚")

print("\n--- Part 2: Lambda è¡¨è¾¾å¼ ---")
# è¿™é‡Œçš„ lambda y: y+y å°±æ˜¯ä¸€ä¸ªä¸éœ€è¦åå­—çš„å‡½æ•°
# é€»è¾‘ï¼šä¼ å…¥ 3 -> 3+3 -> 6 -> 6+6 -> 12
val = twice(lambda y: y+y, 3) 
print(f"Lambda è®¡ç®—ç»“æœ: {val}")
```

```python
# Cell 2: éª°å­æ¨¡æ‹Ÿ (é«˜é˜¶å‡½æ•°)

from random import randint

# å®šä¹‰ä¸¤ä¸ªéª°å­å‡½æ•°
def six_sided():
    return randint(1, 6) # è¿”å›1åˆ°6çš„éšæœºæ•´æ•°

def eight_sided():
    return randint(1, 8)

def repeats(n, dice):
    """
    n: æŠ•æ·æ¬¡æ•°
    dice: è¿™æ˜¯ä¸€ä¸ªå‡½æ•°ï¼æ¯”å¦‚ six_sided
    """
    print(f"--- å¼€å§‹æŠ•æ· {n} æ¬¡ ---")
    count = 0
    previous = None # ç”¨äºè®°å½•ä¸Šä¸€æ¬¡çš„ç‚¹æ•°
    
    while n > 0:
        # è°ƒç”¨ä¼ å…¥çš„å‡½æ•°ï¼Œè·å–ç‚¹æ•°
        outcome = dice()
        
        # æ‰“å°å½“å‰æƒ…å†µ
        print(f"å½“å‰ç‚¹æ•°: {outcome}, ä¸Šä¸€æ¬¡ç‚¹æ•°: {previous}", end=" ")
        
        if previous == outcome:
            print("-> ğŸ”” å‘ç°é‡å¤!")
            count += 1
        else:
            print("-> æ— é‡å¤")
            
        # å…³é”®ä¸€æ­¥ï¼šåœ¨è¿›å…¥ä¸‹ä¸€è½®å‰ï¼ŒæŠŠå½“å‰ç‚¹æ•°å­˜ä¸º"ä¸Šä¸€æ¬¡ç‚¹æ•°"
        previous = outcome
        n -= 1
        
    return count

# è¿è¡Œæµ‹è¯•
total_repeats = repeats(5, six_sided)
print(f"\næœ€ç»ˆç»Ÿè®¡ï¼šå…±æœ‰ {total_repeats} æ¬¡é‡å¤")
```

```python
# Cell 3: é—­åŒ…å·¥å‚

def reprint(n):
    # n åœ¨è¿™é‡Œè¢«é”å®š (ä¾‹å¦‚ n=3)
    
    def a(word):
        # å¿…é¡»æŠŠ n èµ‹å€¼ç»™ä¸€ä¸ªæ–°çš„å±€éƒ¨å˜é‡ k
        # å¦‚æœç›´æ¥ä¿®æ”¹ n (n -= 1)ï¼ŒæŠ¥é”™ï¼Œå› ä¸ºå®ƒè®¤ä¸ºä½ è¯•å›¾ä¿®æ”¹é—­åŒ…åªè¯»å˜é‡
        k = n 
        while k > 0:
            print(f"{word} (å‰©ä½™æ¬¡æ•° {k-1})")
            k -= 1
    
    return a # è¿”å›è¿™ä¸ª"å®šåˆ¶å¥½æ¬¡æ•°"çš„å‡½æ•°

# 1. é€ ä¸€ä¸ª"æ‰“å°3æ¬¡æœº"
print_three_times = reprint(3)

# 2. ä½¿ç”¨å®ƒ
print("--- æµ‹è¯•æ‰“å° 3 æ¬¡ ---")
print_three_times("Hello")

# 3. é€ ä¸€ä¸ª"æ‰“å°2æ¬¡æœº"å¹¶ç«‹åˆ»ä½¿ç”¨
print("\n--- æµ‹è¯•æ‰“å° 2 æ¬¡ ---")
reprint(2)("World")
```

```python
# Cell 4: æ‹†è§£å¤æ‚çš„ Lambda è¡¨è¾¾å¼

# åŸä»£ç : (lambda f: lambda x: f(f(x)))(lambda y: y * y)(3)

print("æ­¥éª¤ 1: å®šä¹‰å·¥å…·å‡½æ•°")
# å®šä¹‰å¹³æ–¹å‡½æ•° (å¯¹åº”åŸä»£ç çš„ lambda y: y * y)
square = lambda y: y * y
print(f"square(3) = {square(3)}")

print("\næ­¥éª¤ 2: å®šä¹‰é«˜é˜¶å‡½æ•°")
# å®šä¹‰"æ‰§è¡Œä¸¤æ¬¡"çš„é€»è¾‘ (å¯¹åº”åŸä»£ç çš„ lambda f: lambda x: ...)

# å®ƒæ¥æ”¶ä¸€ä¸ªå‡½æ•° fï¼Œè¿”å›ä¸€ä¸ªæ–°çš„å‡½æ•°ï¼ï¼ï¼ï¼ï¼ï¼
def make_do_twice(f):
    return lambda x: f(f(x))

print("\næ­¥éª¤ 3: ç»„åˆ")
# æŠŠ square ä¼ ç»™ make_do_twice
do_square_twice = make_do_twice(square)
# ç°åœ¨çš„é€»è¾‘ç­‰åŒäº: lambda x: square(square(x))

print("\næ­¥éª¤ 4: æ‰§è¡Œ")
# ä¼ å…¥ 3
# square(3) -> 9
# square(9) -> 81
result = do_square_twice(3)
print(f"æœ€ç»ˆç»“æœ: {result}")

# éªŒè¯ä¸€è¡Œæµå†™æ³•
print(f"ä¸€è¡ŒæµéªŒè¯: {(lambda f: lambda x: f(f(x)))(lambda y: y * y)(3)}")
```

```python
# Cell 5: Snap ä¸ Chat çš„å˜é‡é­”æ³•

# 1. åˆå§‹å®šä¹‰
# è¿™æ˜¯ä¸€ä¸ªæ¥æ”¶å‚æ•° chat çš„å‡½æ•°ï¼Œè¿”å›ä¸€ä¸ªæ— å‚ lambdaå‡½æ•°
# æ³¨æ„ï¼šè¿”å›å€¼ lambda é‡Œé¢çš„ snap æŒ‡å‘çš„æ˜¯"å…¨å±€å˜é‡ snap"
snap = lambda arg: lambda: snap(arg) 

# 2. è¿™é‡Œçš„ snap(2020) ä½¿ç”¨çš„æ˜¯ä¸Šé¢å®šä¹‰çš„ snap
# å®ƒè¿”å›ä¸€ä¸ª lambda å¯¹è±¡ï¼Œè¿™ä¸ªå¯¹è±¡è®°ä½äº† arg=2020
# æˆ‘ä»¬æš‚æ—¶å«å®ƒ secret_lambda
temp_result = snap(2020) 

print("èµ‹å€¼å‰: snap æ˜¯ lambda, temp_result æ˜¯å†…å±‚ lambda")

# 3. å¹¶è¡Œèµ‹å€¼ (è¿™æ˜¯é­”æ³•å‘ç”Ÿçš„æ—¶åˆ»)
# å…¨å±€å˜é‡ snap å˜æˆäº† print å‡½æ•°
# å…¨å±€å˜é‡ chat å˜æˆäº†ä¸Šé¢çš„ temp_result (é‚£ä¸ªè®°ä½äº†2020çš„ lambda)
snap, chat = print, temp_result

print(f"èµ‹å€¼å: snap ç°åœ¨æ˜¯ {snap.__name__} å‡½æ•°")

# 4. è°ƒç”¨ chat()
# æ‰§è¡Œé€»è¾‘: 
# -> è¿è¡Œ secret_lambda
# -> ä»£ç æ˜¯: return snap(arg)
# -> æŸ¥æ‰¾ snap? -> å®ƒæ˜¯ print!
# -> æŸ¥æ‰¾ arg?  -> å®ƒæ˜¯ 2020!
# -> æ‰§è¡Œ print(2020)
print("\n--- æ‰§è¡Œ chat() ---")
chat() # è¾“å‡º 2020

print("\n--- å†æ¬¡æ‰§è¡Œ chat() ---")
# ç»“æœä¸€æ ·ï¼Œå› ä¸ºé—­åŒ…é‡Œçš„ 2020 æ²¡å˜ï¼Œå…¨å±€çš„ snap (print) ä¹Ÿæ²¡å˜
chat()
```

```python
# ä¸¤ä¸ªç©å®¶è½®æµé€šè¿‡å„è‡ªçš„â€œç­–ç•¥å‡½æ•°â€ (strategy)
#  å¾€æ€»æ•° n ä¸ŠåŠ ä¸€ä¸ªæ•°å­—ï¼ˆé€šå¸¸æ˜¯ 1 åˆ° 3ï¼‰
# ================= PART 1: The Twenty-one Game =================

# 1. æ¸¸æˆä¸»å¾ªç¯å¼•æ“
def play(strategy0, strategy1, goal=21):
    """
    æ¨¡æ‹Ÿæ¸¸æˆè¿‡ç¨‹å¹¶è¿”å›èµ¢å®¶çš„ç´¢å¼•ã€‚
    è§„åˆ™ï¼šè½®æµåŠ æ•°ï¼Œè°è®©æ€»æ•° n è¾¾åˆ°æˆ–è¶…è¿‡ goalï¼Œè°å°±è¾“äº†ï¼ˆè¿”å›å¯¹æ–¹ç´¢å¼•ï¼‰ã€‚
    """
    n = 0
    who = 0  # 0å·ç©å®¶å…ˆæ‰‹
    
    print(f"--- æ¸¸æˆå¼€å§‹ (ç›®æ ‡: {goal}) ---")
    
    while n < goal:
        if who == 0:
            # ç©å®¶0è¡ŒåŠ¨
            added = strategy0(n)
            n = n + added
            # è¡ŒåŠ¨å®Œï¼Œåˆ‡æ¢åˆ°ç©å®¶1
            who = 1
        elif who == 1:
            # ç©å®¶1è¡ŒåŠ¨
            added = strategy1(n)
            n = n + added
            # è¡ŒåŠ¨å®Œï¼Œåˆ‡æ¢åˆ°ç©å®¶0
            who = 0
            
    # å¾ªç¯ç»“æŸæ„å‘³ç€ n >= goalã€‚
    # æ­¤æ—¶ who æŒ‡å‘çš„æ˜¯â€œæ²¡æœ‰è¿›è¡Œæœ€åä¸€æ­¥æ“ä½œçš„äººâ€ï¼Œä¹Ÿå°±æ˜¯èµ¢å®¶ã€‚
    print(f"--- æ¸¸æˆç»“æŸï¼Œæ€»æ•° n={n}ï¼Œèµ¢å®¶æ˜¯ Player {who} ---")
    return who

# 2. åŸºç¡€ç­–ç•¥ï¼šæ°¸è¿œåŠ  2
def two_strat(n):
    return 2

# 3. å®Œç¾ç­–ç•¥ï¼šå°è¯•å‡‘æˆ 4 çš„å€æ•°
def best_strat(n):
    # è¯•å›¾åŠ ä¸Š (4 - n%4)ï¼Œä½†åœ¨ 1åˆ°3 ä¹‹é—´å–å€¼
    # ä¾‹å¦‚ï¼šn=1 (ä½™1), 4-1=3, åŠ 3 -> n=4
    return min(4 - n % 4, 3)

# 4. äº¤äº’å¼ç­–ç•¥ï¼šè®©ä½ è‡ªå·±ç©
def interactive_strat(n):
    choice = 0
    # å¼ºåˆ¶è¾“å…¥ 1 åˆ° 3 ä¹‹é—´çš„æ•´æ•°
    while choice < 1 or choice > 3:
        print(f'å½“å‰æ€»æ•°æ˜¯ {n}ï¼Œä½ è¦åŠ å¤šå°‘ (1-3)?', end=' ')
        try:
            val = input()
            # é˜²æ­¢åœ¨éäº¤äº’ç¯å¢ƒæŠ¥é”™ï¼Œç®€å•å¤„ç†
            if not val: return 1 
            choice = int(val)
        except ValueError:
            pass
    return choice

# 5. è£…é¥°å™¨ï¼šè®©ç­–ç•¥â€œä¼šè¯´è¯â€
def noisy(who, s):
    """è¿”å›ä¸€ä¸ªæ–°çš„å‡½æ•°ï¼Œè¯¥å‡½æ•°ä¼šæ‰§è¡Œç­–ç•¥ sï¼Œå¹¶æ‰“å°è¿‡ç¨‹"""
    def strat(n):
        choice = s(n)
        print(f'Player {who} (ä½¿ç”¨ç­–ç•¥ {s.__name__}) å°† {choice} åŠ åˆ° {n}ï¼Œå˜ä¸º {choice + n}')
        return choice
    return strat

# --- æµ‹è¯•è¿è¡Œ ---

# åœºæ™¯ 1: ä¸¤ä¸ªåªä¼šåŠ  2 çš„æœºå™¨äººäº’æ®´
print(">>> æµ‹è¯• 1: ä¸¤ä¸ª 'two_strat' å¯¹æˆ˜")
winner = play(noisy(0, two_strat), noisy(1, two_strat))
print(f"æµ‹è¯• 1 ç»“æœ: Player {winner} è·èƒœ\n")

# åœºæ™¯ 2: å®Œç¾ç­–ç•¥ vs ç¬¨ç¬¨ç­–ç•¥
print(">>> æµ‹è¯• 2: å®Œç¾ç­–ç•¥(P0) vs ç¬¨ç¬¨ç­–ç•¥(P1)")
# å®Œç¾ç­–ç•¥åº”è¯¥èƒ½èµ¢
play(noisy(0, best_strat), noisy(1, two_strat))
```

```python
# ================= PART 2: Nearest Prime =================

def is_prime(n):
    """åˆ¤æ–­ n æ˜¯å¦ä¸ºè´¨æ•°"""
    if n < 2: return False
    k = 2
    while k < n:
        if n % k == 0:
            return False
        k = k + 1
    return True

def nearest_prime(n):
    """
    è¿”å›ç¦» n æœ€è¿‘çš„è´¨æ•°ã€‚
    å¦‚æœæ˜¯å¹³å±€ï¼ˆè·ç¦»ä¸¤ä¸ªè´¨æ•°ç›¸ç­‰ï¼‰ï¼Œè¿”å›è¾ƒå¤§çš„é‚£ä¸ªã€‚
    """
    k = 0
    # è¿™æ˜¯ä¸€ä¸ªæ— é™å¾ªç¯ï¼Œç›´åˆ°æ‰¾åˆ°è´¨æ•° return ä¸ºæ­¢
    while True:
        candidate = n + k
        
        # è°ƒè¯•æ‰“å°ï¼Œè®©ä½ çœ‹åˆ°æœç´¢è¿‡ç¨‹
        print(f"æ­£åœ¨æ£€æŸ¥ n={n}, åç§» k={k}, å€™é€‰æ•°={candidate}")
        
        if is_prime(candidate):
            return candidate
        
        # æ›´æ–° k çš„é€»è¾‘ï¼šç”Ÿæˆ 0, 1, -1, 2, -2, 3, -3 ... åºåˆ—
        if k > 0:
            k = -k      # æ­£å˜è´Ÿ
        else:
            k = -k + 1  # è´Ÿå˜æ­£ï¼Œä¸”æ­¥é•¿+1

# --- æµ‹è¯•è¿è¡Œ ---

print(f">>> nearest_prime(8) -> æœŸæœ› 7 (è·ç¦»1)")
print(f"ç»“æœ: {nearest_prime(8)}")

print(f"\n>>> nearest_prime(11) -> æœŸæœ› 11 (è·ç¦»0)")
print(f"ç»“æœ: {nearest_prime(11)}")

print(f"\n>>> nearest_prime(4) -> æœŸæœ› 5 (3å’Œ5è·ç¦»éƒ½æ˜¯1ï¼Œå–å¤§çš„)")
# æœç´¢é¡ºåºæ¼”ç¤º: 
# k=0 (4, No) -> k=1 (5, Yes!) -> Return
print(f"ç»“æœ: {nearest_prime(4)}")

print(f"\n>>> nearest_prime(21) -> æœŸæœ› 23")
print(f"ç»“æœ: {nearest_prime(21)}")
```

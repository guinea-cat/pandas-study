```python
# Scipy 高级数学算法 如微积分 统计 信号处理

# 统计模块
# 计算 P 值、正态性检验、生成各种分布（正态、T 分布、泊松分布）。
# 验证模型改进是否显著（T-检验），在特征工程中查看数据分布
from scipy import stats
import numpy as np

# 正态分布随机数 loc=均值 scale=标准差 size=数量
# Random Variates 随机变量样本
data = stats.norm.rvs(loc=0, scale=1, size=1000)

print(stats.describe(data))
print(f"均值: {np.mean(data)}, 标准差: {np.std(data)}")
print(f"偏度: {stats.skew(data)}, 峰度: {stats.kurtosis(data)}")

# 法二 f-string 必须在同一行 三引号可以换行
res = stats.describe(data)

print(f"\n法二 均值: {res.mean:.2f}\n")
print(f"""法二 nobs={res.nobs}, minmax=({res.minmax[0]:.2f}, 
      {res.minmax[1]:.2f}), mean={res.mean:.2f}""")

# T-检验：比较两组样本 均值是否有显著差异
group1 = [1, 2, 3, 4, 5]
group2 = [1.2, 2.2, 3.2, 4.2, 5.2]
t_stat, p_val = stats.ttest_ind(group1, group2)
print(f"P值: {p_val:.2f}") # 如果 p < 0.05，说明差异显著


```

```python
# 优化模块 机器学习的本质就是“损失函数最小化”
from scipy import optimize
def objective_function(x):
    return (x - 3) ** 2 + 2
result = optimize.minimize(objective_function, x0=0) # x0 初始猜测值

print(f"最优解 x: {result.x[0]:.2f}, 最小值: {result.fun:.2f}")
print(type(result.x)) # 最优解数组 result.fun是最优解处的取值
print()


from scipy.optimize import curve_fit

# 要拟合的模型函数 
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

# 有一堆散点数据
xdata = np.linspace(0, 4, 50) # 在 0-4 之间均匀生成 50 个点
y = func(xdata, 2.5, 1.3, 0.5)
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise

# np.random.normal(loc=170,scale=5,size=100) 
# 使用 curve_fit（最小二乘法) 找到最佳参数 a, b, c
# popt 是拟合出的参数数组， pcov 是参数协方差矩阵，对角元可算参数标准差
popt, pcov = curve_fit(func, xdata, ydata)
print(f"拟合出的参数: {popt}")
print(f"拟合出的 a={popt[0]:.2f}, b={popt[1]:.2f}, c={popt[2]:.2f}")
print('参数的协方差矩阵:\n', pcov)

import matplotlib.pyplot as plt

plt.scatter(xdata, ydata, label='Data (with noise)') # 画散点
plt.plot(xdata, func(xdata, *popt), 'r-', label='Fitted Curve') # 画拟合曲线
plt.legend()
plt.show()
```

```python
# 线性代数模块 深度学习的基础
# 常见功能：矩阵分解（SVD, QR, Schur）、求解线性方程组。
# 在大模型的参数压缩（Quantization）或权重分析中，常用SVD来降维
from scipy import linalg

matrix = np.array([[1, 2], [3, 4]])

inv_matrix = linalg.inv(matrix)
print(f"矩阵的逆:\n{inv_matrix}")

# 奇异值分解 (SVD) - 大模型压缩
U, s, Vh = linalg.svd(matrix)
print(f"奇异值: {s}")
```

```python
# 稀疏矩阵模块 节省内存
# 处理大规模文本数据（如词袋模型、TF-IDF）时，矩阵中绝大多数元素都是 0
# 处理 NLP 竞赛数据时，普通NumPy 矩阵会内存爆炸，必须用稀疏矩阵

from scipy import sparse
# 创建一个 4x4 的稀疏矩阵，只有少数元素非零
sparse_matrix = sparse.csr_matrix([[0, 0, 3, 0],
                                   [4, 0, 0, 0],
                                      [0, 0, 0, 5],
                                      [0, 6, 0, 0]])
print("稀疏矩阵:\n", sparse_matrix)
print("稀疏矩阵转换为密集矩阵:\n", sparse_matrix.todense())
```

```python
from scipy.sparse import csr_matrix
# 1. 坐标定位：在这些坐标点上有非零值
row  = np.array([0, 0, 1, 2]) # 对应的行号
col  = np.array([0, 2, 2, 0]) # 对应的列号

# 2. 具体数值：在上述坐标点填入的值
data = np.array([1, 2, 3, 4]) 

# 3. 组合：(0,0)位置填1，(0,2)位置填2，(1,2)位置填3，(2,0)位置填4 zip
sparse_mtx = csr_matrix((data, (row, col)), shape=(5, 5))
print("构建的稀疏矩阵:\n", sparse_mtx.todense())
print()
print(f'稀疏矩阵非零元位置及数值为：\n{sparse_mtx}')
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 0, 1, 0])

f_linear = interp1d(x, y, kind='linear')  # 线性插值（折线）
f_cubic = interp1d(x, y, kind='cubic')    # 三次样条插值（平滑曲线）

x_new = np.linspace(0, 4, 50)
y_linear = f_linear(x_new) # 计算线性插值
y_cubic = f_cubic(x_new)

# --- 方法一：使用 Matplotlib 绘图 ---
plt.figure(figsize=(10, 5))

# 画出原始散点 zorder越大，图层越不容易被遮住
# 线条（plot）是 2，散点（scatter）是 3
plt.scatter(x, y, color='red', label='Original Points', zorder=5)
# 画出线性插值（折线）
plt.plot(x_new, y_linear, '--', label='Linear Interpolation', alpha=0.6)
# 画出三次插值（平滑曲线）alpha=1不透明，实心
plt.plot(x_new, y_cubic, label='Cubic Spline (Smooth)', linewidth=2)

plt.title("Matplotlib: Interpolation Comparison")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6) # ：点线 - 实线
plt.show()

# --- 方法二：使用 Seaborn 绘图 ---
# Seaborn 通常配合 Pandas 使用，但也可以直接接受数组
sns.set_theme(style="whitegrid") 
plt.figure(figsize=(10, 5))

# 用 lineplot 画平滑曲线
sns.lineplot(x=x_new, y=y_cubic, label="Smooth Curve", color="teal")
# 用 scatterplot 画原始点
sns.scatterplot(x=x, y=y, color="red", s=100, label="Raw Data")

plt.title("Seaborn: Smooth Curve Visualization")
plt.show()
```

```python
from scipy.spatial import distance # 空间距离
# 计算大模型 Embedding（词向量）之间的 余弦相似度
# 两个句子的向量表示
vec_a = [1, 2, 3] # 你好
vec_b = [1, 2, 3.1] # 您好

# 欧几里得距离 (直线距离)
dst_euclidean = distance.euclidean(vec_a, vec_b)

# 计算余弦距离 (大模型中最常用的相似度指标) 向量夹角 方向是否一致
# 文字处理中，一个句子长一个句子短没关系，只要意思（方向）对
dst_cosine = distance.cosine(vec_a, vec_b)

print(f"欧式距离: {dst_euclidean:.4f}")
print(f"余弦距离: {dst_cosine:.4f}") # 越接近 0 表示越相似
```

```python
# 信号平滑与过滤 如用于去除时间序列数据中的噪声
# Medfilt (中值滤波)
from scipy import signal
import matplotlib.pyplot as plt

# 模拟带噪声的信号
t = np.linspace(0, 1, 100) # 时间轴
x = np.sin(2 * np.pi * 5 * t)  # 纯净信号
x_noise = x + np.random.normal(0, 0.5, 100) # 加了噪声

# 使用中值滤波平滑信号（核大小为 5）窗口越大越平缓
# 滑动窗口 套住信号中5个连续点 取中位数替换窗口中心原有值
# 噪声（比如突然跳出来的一个极高值）通常是孤立的。
# 在排序后，极值会被排在两端，而中位数通常代表了真实的信号趋势
x_smoothed = signal.medfilt(x_noise, kernel_size=5)

plt.figure(figsize=(10, 4))
plt.plot(t, x_noise, label='Noisy Signal', alpha=0.5)
plt.plot(t, x_smoothed, label='Smoothed Signal', linewidth=2)
plt.legend()
plt.title("Signal Smoothing with SciPy")
plt.show()
```

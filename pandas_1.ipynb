```python
# 创建 df4：数据类型包括 float, datetime, object, int
df4 = pd.DataFrame({
    "A": [1.0, np.nan, 3.5, 4.0],
    "B": [pd.Timestamp("20230101"), pd.Timestamp("20230102"), np.nan, pd.Timestamp("20230104")],
    "C": ["foo", "bar", np.nan, "baz"],
    "D": [10, 20, 30, np.nan],
    "E": [np.nan, 1.2, np.nan, 4.5]
})

print("--- 原始的 df4 表格（含有缺失值 NaN） ---")
print(df4)
print()
# 检测缺失值
print(df4.isnull())  # 返回布尔值表格   
print()
print(df4.isnull().sum())  # 每列缺失值个数
print()
# 填充缺失值
df4_filled_1 = df4.fillna(method='ffill')  # 前向填充
print("--- 用前向填充法填充缺失值后的 df4 表格 ---")
print(df4_filled_1)
print()
df4_filled_2 = df4.fillna(method='bfill')  # 后向填充
print("--- 用后向填充法填充缺失值后的 df4 表格 ---")
print(df4_filled_2)
df4.filled_3 = df4.fillna(df4.mean(numeric_only=True))  # 用均值填充数值列缺失值
print("--- 用均值填充数值列缺失值后的 df4 表格 ---")
print(df4.filled_3)
print()
print(df4.fillna(value=0))  # 用0填充缺失值
print()
```

```python
# NaN 的类型是浮点数。如果你的一列整数里出现了一个缺失值，
# Pandas 会把整列整数强制转换成浮点数（1 变成 1.0）来容纳 NaN
# NaN 和 NaT 都属于浮点数类型 缺失值
print(df4.dtypes)
print(df4)
print()
df4_mean_filled = df4.fillna(df4.mean(numeric_only=True)) 
print(df4_mean_filled)
print()
df4_median_filled = df4.copy()
df4_median_filled['D'] = df4['D'].fillna(df4['D'].median())
df4_median_filled['E'] = df4['E'].fillna(df4['E'].median())
df4_median_filled['A'] = df4['A'].fillna(df4['A'].median())
print("--- 用中位数填充数值列缺失值后的 df4 表格 ---")
print(df4_median_filled)
```

```python
df4_dropped = df4.dropna()  # 删除含有缺失值的行 how='any' 默认值
print("--- 删除含有缺失值的行后的 df4 表格 ---")
print(df4_dropped)
# describe() 会自动忽略缺失值 NaN
print(df4.describe())
print()
print(df4.mean(numeric_only=True))

print("\n--- 每一列非空值的数量 ---")
print(df4.count())
```

```python
df5 = pd.DataFrame({
    "A": [1.0, 1.0, np.nan, 4.0],
    "B": [pd.Timestamp("20230101"), pd.Timestamp("20230101"), np.nan, pd.Timestamp("20230104")],
    "C": ["foo", "foo", np.nan, "baz"],
    "D": [10, 10, np.nan, 40],
    "E": pd.Categorical(["test", "test", np.nan, "train"])
})

print("--- 原始 df5（含有各种缺失值） ---")
print(df5)
print()
print("--- 每列的众数 df ---")
print(df5.mode()) # df
print()
modes = df5.mode().iloc[0]  # 获取每列的众数
print("--- 每列的众数 Series ---")
print(modes)
print()
df5_filled_mode = df5.fillna(modes)  # 用众数填充缺失值
print("--- 用众数填充缺失值后的 df5 表格 ---")
print(df5_filled_mode)
print()

print("--- 只选择 object 和 category 类型的列 df---")
print(df5.select_dtypes(include=['object', 'category']))
print()
# 只选择 object 和 category 类型进行 众数填充
obj_cols = df5.select_dtypes(include=['object', 'category']).columns
df5[obj_cols] = df5[obj_cols].fillna(df5[obj_cols].mode().iloc[0])

# 对数字列 用平均值填充
num_cols = df5.select_dtypes(include=['number']).columns
df5[num_cols] = df5[num_cols].fillna(df5[num_cols].mean())

print("\n--- 混合填充后的结果（分类用众数，数字用平均值） ---")
print(df5)
```

```python
# Groupby df.groupby('分组列')['数值列'].统计函数()
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob'],
    'City': ['北京', '北京', '上海', '上海', '北京', '上海'],
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb'],
    'Sales': [100, 200, 150, 300, 120, 250],
    'Profit': [10, 20, 15, 35, 12, 28]
}
df8 = pd.DataFrame(data)
print("原始 df8:")
print(df8)

print("\n--- 按 Name 分组，计算 Sales 总和 ---")
sales_sum = df8.groupby('Name')['Sales'].sum()
print(sales_sum)

print("\n--- 按 City 分组，计算 Profit 平均值 ---")
profit_mean = df8.groupby('City')['Profit'].mean()
print(profit_mean)
```

```python
print("--- 按 Month 分组，计算 Sales 和 Profit 的总和 ---") # 2列
sales_profit_sum = df8.groupby('Month')[['Sales', 'Profit']].sum()
print(sales_profit_sum)

print("\n--- 按 Name 和 City 分组，计算 Sales 总和 ---")
sales_sum_multi = df8.groupby(['Name', 'City'])['Sales'].sum()
print(sales_sum_multi)

print("\n--- 按 Name和 City 分组，计算 Sales 和 Profit 的均值和总和 ---")
sales_profit_agg = df8.groupby(['Name', 'City'])[['Sales', 'Profit']].agg(['mean', 'sum'])
print(sales_profit_agg)
```

```python
print("按城市查看销售额/利润 的总和和平均值:")
print(df8.groupby('City')[['Sales','Profit']].agg(['sum', 'mean']))

print("\n--- 按 Month 分组，计算 Sales 和 Profit 的总和 ---") # 2列
sales_profit_sum = df8.groupby('Month')[['Sales', 'Profit']].sum()
print(sales_profit_sum)

print()
result = df8.groupby('City').agg({
    'Sales': ['sum', 'mean'],
    'Profit': ['sum', 'mean']
})
print(result)

print("\n对不同列 应用不同统计函数：")
print(df8.groupby('City').agg({
    'Sales': ['sum', 'mean'],
    'Profit': ['sum', 'mean']
}))

print("\n使用自定义字典聚合函数：")
rules = {
    'Sales': 'sum',
    'Profit': 'max'
}
result = df8.groupby('Name').agg(rules)
print(result)
```

```python
# 分组的列 默认变成行索引，可以用 as_index=False 使之保持为普通列
print(df8.groupby('Name', as_index=False)['Sales'].sum())
print()
print(df8.groupby('City', as_index=False).agg({
    'Sales': ['sum', 'mean'],
    'Profit': ['sum', 'mean']
}))

print('\n按 City 分组，计算 Sales 的均值和 Profit 的最大值:')
print(df8.groupby('City', as_index=False).agg({
    'Sales': 'mean',
    'Profit': 'max'
}))

print('\n按 City 和 Name 分组，计算 Sales 的均值和 Profit 的最大值:')
print(df8.groupby(['City','Name'], as_index=False).agg({
    'Sales': 'mean',
    'Profit': 'max'
}))
```

```python
df_left = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

# 
df_right = pd.DataFrame({
    'ID': [1, 2, 5],
    'Sales': [100, 200, 500]
})
print("左表 df_left:")
print(df_left)
print()
print("右表 df_right:")
print(df_right)
```

```python
# pd.concat默认上下拼接，按行拼接，把今天的数据追加到昨天的后面
df_concat = pd.concat([df_left, df_right], axis=0, ignore_index=True)
print("\n--- 按行拼接 df_left 和 df_right ---")
print(df_concat)
print()
# axis = 1 按列拼接 左右拼接 新列强行加到右边
df_concat_col = pd.concat([df_left, df_right], axis=1)
print("\n--- 按列拼接 df_left 和 df_right ---")
print(df_concat_col)
```

```python
# pd.merge：按某 列 的值 对齐合并 merge的简化版为基于index的join
# 根据两张表里共同拥有的那一列（比如 ID），把对应的数据找出来拼在一起
# 默认inner 只保留两张表都有的ID，left以左表为准，右表没数据的填NaN
# right以右表为准，左表没数据的填NaN，outer取并集，缺失值填NaN
df_merged = pd.merge(df_left, df_right, on='ID', how='inner')
print("\n--- 按 ID 列对齐合并 df_left 和 df_right（内连接）---")
print(df_merged)

df_merged_left = pd.merge(df_left, df_right, on='ID', how='left')
print("\n--- 按 ID 列对齐合并 df_left 和 df_right（左连接）---")
print(df_merged_left)

df_merged_outer = pd.merge(df_left, df_right, on='ID', how='outer')
print("\n--- 按 ID 列对齐合并 df_left 和 df_right（外连接）---")
print(df_merged_outer)

df_merged_right = pd.merge(df_left, df_right, on='ID', how='right')
print("\n--- 按 ID 列对齐合并 df_left 和 df_right（右连接）---")
print(df_merged_right)
```

```python
# df.join 默认按照行索引index合并，语法为 df1.join(df2, how='left')

# 先把 ID 设为索引
left_idx = df_left.set_index('ID')
right_idx = df_right.set_index('ID')

print("\n--- 左表 left_idx（ID 设为索引）---")
print(left_idx)
print("\n--- 右表 right_idx（ID 设为索引）---")
print(right_idx)

print("\n--- 使用 join 方法按 ID 列对齐合并 df_left 和 df_right（默认左连接）---")
df_joined = left_idx.join(right_idx, how='left')
print(df_joined)

print("\n--- 使用 join 方法按 ID 列对齐合并 df_left 和 df_right（内连接）---")
print(left_idx.join(right_idx, how='inner'))

print("\n--- 使用 join 方法按 ID 列对齐合并 df_left 和 df_right（外连接）---")
print(left_idx.join(right_idx, how='outer'))

print("\n--- 使用 join 方法按 ID 列对齐合并 df_left 和 df_right（右连接）---")
print(left_idx.join(right_idx, how='right'))
```
